```python
df.Precip=df.Precip.replace("T",0).astype(float)
df.Precip=df.Precip.fillna(df.Precip.mean())
```

```python
df.Snowfall.value_counts()
```

```python
df.Snowfall=df.Snowfall.replace("#VALUE!",np.nan)
df.Snowfall=df.Snowfall.astype(float)
df.Snowfall=df.Snowfall.fillna(df.Snowfall.mean())
```

```python
df.info()
```

```python
df.PRCP.value_counts()
```

```python
df.PRCP=df.PRCP.replace("T",0).astype(float)
df.PRCP=df.PRCP.fillna(df.PRCP.mean())
```

```python
df.MAX = df.MAX.fillna(df.MAX.mode()[0])
df.MIN = df.MIN.fillna(df.MIN.mode()[0])
df.MEA = df.MEA.fillna(df.MEA.mean())
```

```python
df.SNF.value_counts()
```

```python
df.SNF=df.SNF.replace("T",0).astype(float)
df.SNF=df.SNF.fillna(df.SNF.mean())
```

```python
df.drop("Date", axis=1, inplace=True)
```

```python
df.info()
```

```python
df.drop(labels=['MAX','MIN','MEA','SNF'],axis= 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
df.info()
```

```python
df.head()
```

```python
df_mean_temp=df.copy()
```

```python
X=df_mean_temp.drop("MeanTemp",axis=1)
y=df_mean_temp["MeanTemp"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=15)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
regression = LinearRegression()
regression.fit(X_train_scaled, y_train)
```

```python
y_pred = regression.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("r2score:",r2)
print("mae:",mae)
print("mse:",mse)

plt.scatter(y_test,y_pred)
plt.title("Linear Regression")
plt.show()
```

```python
lasso = Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred = lasso.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("r2score:",r2)
print("mae:",mae)
print("mse:",mse)

plt.scatter(y_test,y_pred)
plt.title("Lasso")
plt.show()
```

```python
ridge = Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred = ridge.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("r2score:",r2)
print("mae:",mae)
print("mse:",mse)

plt.scatter(y_test,y_pred)
plt.title("Ridge")
plt.show()
```

```python
elastic = ElasticNet()
elastic.fit(X_train_scaled,y_train)
y_pred = elastic.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("r2score:",r2)
print("mae:",mae)
print("mse:",mse)

plt.scatter(y_test,y_pred)
plt.title("ElasticNet")
plt.show()
```

```python
from sklearn.linear_model import LassoCV,RidgeCV,ElasticNetCV
lassocv = LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
y_pred = lassocv.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("r2score:",r2)
print("mae:",mae)
print("mse:",mse)

plt.scatter(y_test,y_pred)
plt.title("LassoCV")
plt.show()
```

```python
ridgecv = RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train)
y_pred = ridgecv.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("r2score:",r2)
print("mae:",mae)
print("mse:",mse)

plt.scatter(y_test,y_pred)
plt.title("RidgeCV")
plt.show()
```

```python
elasticnetcv = ElasticNetCV(cv=5)
elasticnetcv.fit(X_train_scaled,y_train)
y_pred = elasticnetcv.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("r2score:",r2)
print("mae:",mae)
print("mse:",mse)

plt.scatter(y_test,y_pred)
plt.title("ElasticNetCV")
plt.show()
```
